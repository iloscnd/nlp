```python
words = {}


for line in open("../poleval_corpus/poleval_2grams.txt"):
    _, word1, word2 = line.split()
    l = len(word1)
    if l not in words:
        words[l] = set()
    words[l].add(word1)
    l = len(word2)
    if l not in words:
        words[l] = set()
    words[l].add(word2)
        
            
```

```python

def edit_distance(a,b):
    if len(a) > len(b):
        a, b = b, a
        
    if len(b) == 0:
        return len(a)
    
    prev = range(len(b) + 1)
    for i, c1 in enumerate(a):
        row = [i + 1]
        for j, c2 in enumerate(b):
            row.append(min(prev[j+1] + 1, row[j] + 1, prev[j] + (c1 != c2)))
        prev = row
    
    return prev[-1]
    
    
def compare(a, b):
    if len(a) > len(b):
        a , b= b,a
    
    cost = 0
    for i in range(len(a)):
        cost += edit_distance(a[i], b[i])
    
    for i in range(len(a), len(b)):
        cost += (len(b[i]))
    #print(a,b)
    return cost/len(b)
    
    
def maxMatchRe(string):
    
    if len(string) == 1:
        return [string]
    
    if len(string) == 0:
        return []
    
    for key in sorted(words.keys(), reverse = True):
        
        if len(string) < key:
            continue
        
        if string[:key] in words[key]:
            result = maxMatchRe(string[key:])
            if result == None:
                continue
            result.append(string[:key])
            return result
        
    return None

def maxMatch(string):
    ret = maxMatchRe(string)
    if ret == None:
        return None
    else:
        return list(reversed(ret))
```
