```python

bigrams = {}
classes = {}


it = 0

for line in open("../poleval_corpus/poleval_2grams.txt"):
    num, fst, snd = line.split()
    
    if len(fst) <= 3:
        fst_c = fst
    else:
        fst_c = fst[-3:]
        
    if len(snd) <= 3:
        snd_c = snd
    else:
        snd_c = snd[-3:]
    
    if fst not in classes:
        classes[fst] = {}
    
    if snd not in classes[fst]:
        classes[fst][snd] = 0
    
    classes[fst][snd] += int(num)
    
    if int(num) < 2:
        continue
    
    
    if fst not in bigrams:
        bigrams[fst] = {}
        
    bigrams[fst][snd] = int(num)
    
    
    it += 1
    if it % 1000000 == 0:
        print(it, end=" " )

print("set_loaded")
print(len(bigrams))

```

```python
words = ['judyta dała wczoraj stefanowi czekoladki .'.split()]
words.append('babuleńka miała dwa rogate koziołki .'.split())
words.append('wczoraj wieczorem spotkałem pewną piękną kobietę .'.split())
words.append('jestem głodny , więc idę jeść .'.split())
words.append('nie mam pomysłu na zdania .'.split())
words.append('chyba już wystarczy zdań ...'.split())
```

```python
import itertools

for k,sen in enumerate(words):
    res = []
    for perm in itertools.permutations(sen):
        cost = 0
       
        for i in range(1, len(sen)):
            fst = perm[i-1]
            snd = perm[i]
            
            if len(fst) <= 3:
                fst_c = fst
            else:
                fst_c = fst[-3:]

            if len(snd) <= 3:
                snd_c = snd
            else:
                snd_c = snd[-3:]

            if fst in bigrams:
                cost -= bigrams[fst].get(snd, 0)
               # print(fst, snd, bigrams[fst].get(snd,0))
            
            if fst_c in classes:
                cost -= classes[fst_c].get(snd_c, -20*len(classes[fst_c]))/len(classes[fst_c])
            else:
                cost -= -40
        #print(cost, perm)
        res.append((cost, perm))
    for cost, perm in sorted(res):
        print(-cost, " ".join(perm))
    print()
```
